## <u> T-uplet en Python </u>

Un t-uplet est une séquence <font color='green'>immuable </font> d'éléments. On retrouve avec les objets de type tuple une grande partie des propriétés communes aux séquences d'éléments en Python (chaînes et tableaux).

### <u>1. Définir un tuple directement</u>
On définit un tuple de façon directe en écrivant son contenu entre parenthèses. Executer les exemples suivants et remarquez surtout la difference entre  a = (2,)  et b= (2)

```python
#Exemple 2
a = (2, 3, 4)
# ou
a= 2, 3, 4
print (b)

#Cas particuliers:

# 1. Pour définir un tuple vide, on peut uniquement affecter un couple de parenthèses :
a = ()
print (type(a))

# 2. Pour éviter les confusions, il est aussi possible de forcer le typage grâce à la fonction tuple() :
a = tuple()
print (type(a))

# 3. Pour définir un tuple ne contenant qu'un seul élément (simplet), il faut faire suivre cet élément par une virgule dans les parenthèses :
a = (2,)
b= (2)
print (type(a), type (b))

```

### <u> 2. Définir un tuple par comprehension </u>
On peut aussi définir un T-uplet par compréhension. C'est la meme syntaxe utilisée dans les tableaux a une différence près:
* Il faut encapsuler la définition entre parenthèses.
* Faire précéder la parenthèse par le mot clé tuple

Exécuter l'exemple suivant :


```python
#Exemple 3 :
# creer par comprehension un t-uplet  qui contient les entier de -40 a 40 inclus.
t= tuple(i for i in range (-40,41))
print (t)

#creer par comprehension un t-uplet qui contient tous les nombres pairs  entre 0 et 100, divisibles par 3.
t2= tuple (i for i in range(0,101)  if i%2==0 and i%3==0)
print(t2)

```

### <u>  6. Parcourir les éléments d'un t-uplet </u>
comme pour les chaine de caracteres et les listes, il existe trois méthodes pour parcourir les valeurs d'un t-uplet.

- On peut parcourir élément par élément d'un t-uplet avec for ... in ...:
- On peut parcourir le t-uplet indice par indice grâce à for ... in range(len(...)):
- On peut aussi parcourir le t-uplet en utilisant à la fois les indices et les valeurs des éléments grâce à for ... in enumerate(...):

### Exemples:

```python
# Exemple  8: Parcourir un tuple:
a=(3,4,6,7,-8,10)
for element in a:
    print (element)

for i in range(len(a)):
    print (a[i])

for index, valeur  in enumerate(a):
    print(index, valeur)

```

```python
# Exemple 9 : faire cet exemple.
# on construit les tuples par concatenation comme les chaines de caracteres.
'''
Complétez la définition de la fonction signes() qui prend pour paramètre un t-uplet t
contenant des eniers et renvoie un t-uplet de même longueur contenant un 0 lorsque l'élément
correspondant de t est strictement positif, un 1 sinon.
'''
def signes(t):

    """
    t – tuple, t-uplet d’entiers
    Sortie: tuple – t-uplet de même longueur que t contenant :
            0 si l’élément de même rang dans t est strictement positif
            1 si l’élément de même rang dans t est négatif
    >>> signes( (3, -2, -6, 0) )
    (0, 1, 1, 1)
    """
    t2=tuple()
    for element in t:
        if element>0:
            t2=t2+(0,)
        else:
            t2=t2+(1,)
    return t2

print(signes((4,-2,-1,25,-14,2,24,-2)))
        

```

### <u> 9.  Test d'appartenance, comparaison </u>
On peut tester l'appartenance d'un élément à un tuple avec l'opérateur <font  color= 'red' >in </font> et comparer deux tuples avec ==. La comparaison de deux tuples s'effectue de gauche à droite

```python
# Exemple 12
b = (3,4,6,89,30,92)
c = (3,4,6,89,30,92)

# essayer l'instruction suivante:
print (3 in b) # cela devrait nous donner True, car 3 est dans le tuple b
print (110 in b ) # cela devrait nous donner False, car 101 n'est pas dans le tuple b
print (100 not in b) # cela devrait nous donner True
print (b==c) # oui cela est vrai
# par contre :
print(id(b)==id(c)) # nous donnera une valeur False.

```

### <u> 10.  Recherche d'elements dans un tuple </u>
 Comme dans les listes, on peut rechercher l'index de la première occurence d'un élément dans un tuple avec la méthode index, ou compter le nombre d'occurences d'une valeur avec count.

```python
#Exemple 13: executer ces lignes de codes
from random import randint
e = tuple([randint(1,6) for _ in range(10)])
print (e.index(5)) # cette instruction donne l'indice de la valeur 5 si 5 est dans e
print (e.index(1)) # cette instruction donne l'indice de la valeur 1 si 1 est dans e
```
